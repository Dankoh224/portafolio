```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python
def chatbot():
  print('Hola, soy ChatboT, tu fiel servidor, ¿en que puedo ser útil?')
  pairs = [
      [r'(.*)_____(.*)',['_____']],
      [r'(.*)_____(.*)',['_____']],
      [r'(.*)_____(.*)',['_____']],
      [r'(.*)_____(.*)',['_____']]
  ]
  chat = Chat(pairs,reflections)
  chat.converse()
```

```python
if __name__ == '__main__':
  nltk.download('punkt')
  chatbot()
```
