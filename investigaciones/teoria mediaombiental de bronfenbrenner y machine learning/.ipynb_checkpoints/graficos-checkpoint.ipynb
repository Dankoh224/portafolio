 ### 3.6.1 Separar la variable repitencia.

##### La variable predictora en este proyecto es repitencia. Para realizar comparaciones, hay que separar esta variable en:
##### 1.- Los que repiten el año escolar (1).
##### 2.- Los que NO repiten el año escolar (0).

#### Porcentaje de estudiantes que repetirán el año escolar:

```python
# Contamos el número de alumnos que repiten y los que no.
# Estamos ante un conjunto de datos desequilibrado (muchos No repiten, pocos repiten).
repitentes = estudiantes[estudiantes['repitencia'] == 1]
no_repitentes = estudiantes[estudiantes['repitencia'] == 0]
print("El total de estudiantes es {}, "
      "la cantidad de alumnos repitentes son {} "
      "y los que avanzarán al siguiente año escolar son {}.".format(len(estudiantes),len(repitentes),len(no_repitentes)))

print("Por lo tanto, un {:.1f}% de los alumnos, "
      "repiten el año escolar.".format(len(repitentes)*100/len(estudiantes)))
```

## 5.1 Regresión logística para la predicción de repitencia.

### 5.1.1 Entrenamiento del modelo de regresión logística.

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicción
y_pred = classifier.predict(X_test)
```

```python
# Revisamos
y_pred
```

### 5.1.2 Evaluación del modelo.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# ACCURACY
accuracy_score(y_test,y_pred)
```

```python
# MATRIZ DE CONFUSIÓN 
def plot_matriz(y_test,y_pred):
  cm = confusion_matrix(y_test, y_pred)
  sns.set(font_scale=1)
  sns.heatmap(cm, annot=True,cmap='Blues',fmt='g')
  plt.title("Matriz de confusion")
  plt.xlabel("Valores predichos")
  plt.ylabel("Valores reales")
  plt.show()
# Graficar
plot_matriz(y_test,y_pred)
```

```python
# CLASSIFICATION REPORT
print(classification_report(y_test,y_pred))
```
