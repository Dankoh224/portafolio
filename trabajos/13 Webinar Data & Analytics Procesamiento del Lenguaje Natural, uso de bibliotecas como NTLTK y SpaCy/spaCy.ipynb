#3. LEMATIZACIÓN.

```python
import pandas as pd
```

```python
texto_2 = "una vez conocí dos veces a dos ladrones que robaban por allá, robaban aquí, robaron acúa, dejaron de robar y volvieron robando más que en cualquier robo que se haya robado antes"
```

```python
documento_2 = nlp(texto_2) # creamos otra instancia de clase documento mediante nlp y el texto_2
```

```python
print(documento_2)
```

```python
# creamos un for para imprimir el token y su lematización.
for token in documento_2:
  if token.is_punct or token.is_stop or token.is_space:
    continue
  print(token)
  print(token.lemma_)
  print(' - ')
```

```python
# bucle para crear una lista con los resultados anteriores:
lista_token_lemmas = []
for token in documento_2:
  if token.is_punct or token.is_stop or token.is_space:
    continue
  lista_token_lemmas.append([token.lower_, token.lemma_.lower()])
lista_token_lemmas
```

```python
# creamos el dataframe
token_lemmas_df = pd.DataFrame(lista_token_lemmas, columns = ['token','lemma'])
token_lemmas_df
```

```python
# y hacemos un agrupado por token para contar cuantos elementos existen:
agrupado = token_lemmas_df.groupby('token')['token'].count()
agrupado.sort_values(ascending = False, inplace = True)
agrupado
```

```python
# hacemos lo mismo para contar la cantidad de palabras que se encuentran en lemma:
agrupado_2 = token_lemmas_df.groupby('lemma')['lemma'].count()
agrupado_2.sort_values(ascending = False, inplace = True)
agrupado_2
```
